## Replicate the split-apply-combine without groupby

```python
def get_min_price(df):
    return pd.DataFrame({'min_price': [df['price'].min()]})

min_prices1 = []
for make in df['make'].unique():
    make_df = df[df['make'] == make]
    min_prices2 = make_df.groupby('make').apply(get_min_price)
    min_prices1.append(min_prices2)

min_prices2 = pd.concat(min_prices1)
```

```python
min_prices2
```

## Create 2 separate dictionaries and join them as a new dataset

```python
dict1 = {'make': ['Honda', 'Toyota', 'Audi', 'Bmw'], 'fuel_type': ['gas', 'gas', 'gas', 'gas'], 'num_doors': [4, 4, 4, 2]}
dict2 = {'make': ['Honda', 'Toyota', 'Audi', 'Bmw'], 'drive_wheel': ['fwd', 'fwd', 'fwd', 'rwd'], 'body_style': ['sedan', 'sedan', 'wagon', 'sedan']}

df1 = pd.DataFrame.from_dict(dict1)
df2 = pd.DataFrame.from_dict(dict2)

new_df = pd.merge(df1, df2, on='make')
```
